```python
scaler = preprocessing.StandardScaler()
scaler.fit(x)
scaler.transform(x)
```

```python
def cost(x,y,m,c):
    summ=0
    for i in range(x.shape[0]):
        summ+= (y[i]-((m*x[i]).sum()+c))**2
    return summ/x.shape[0];
```

```python
def train(x_train,y_train,m,c,alpha):
    m_slope = np.zeros(m.size)
    c_slope=0
    summ=0
    
    for j in range(x_train.shape[1]):
        for i in range(x_train.shape[0]):
            summ+=2*(y[i]-((m*x_train[i]).sum()+c))*(-1*x_train[i][j])
        m_slope[j]=summ/x_train.shape[0]
      
    summ=0
    
    for i in range(x_train.shape[0]):
        summ+=2*(y[i]-((m*x_train[i]).sum()+c))*(-1)
    c_slope = summ/x_train.shape[0]
    
    m = m-m_slope*alpha
    c = c-c_slope*alpha
    
    return m,c
            
```

```python
def help(x_train,y_train,no_of_iteration,alpha):
    m = np.zeros(x_train.shape[1])
    c = 0
    for i in range(no_of_iteration):
        m,c = train(x_train,y_train,m,c,alpha)
        print(cost(x_train,y_train,m,c)," ")
    return m,c
```
